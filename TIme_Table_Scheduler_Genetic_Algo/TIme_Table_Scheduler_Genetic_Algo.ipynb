```python
import pandas as pd
import random
from numpy import random as rn


'''
GROUP MEMBERS:
Talha Mustafa (18I-0573)
Tayyab Ali (18I-0531)

SECTION: D
'''

"""---------------------------Global Variable Declarations and Initializations---------------------------"""
# Initial population size
population_size = 50
crossover_probability = round(rn.uniform(low=0.3, high=1.0), 1)
mutation_probability = round(rn.uniform(low=0.0, high=0.5), 1)

# 'Data/' for data generated by us
# 'Data1/' for provided data
folder = 'Data/'

if folder == 'Data1/':
    # 2 WEEKS FOR EXAMS
    days = [("Monday", "01/05/2021"), ("Tuesday", "02/05/2021"), ("Wednesday", "03/05/2021"), ("Thursday", "04/05/2021"),
            ("Friday", "05/05/2021"), ("Monday", "08/05/2021"), ("Tuesday", "09/05/2021"), ("Wednesday", "10/05/2021"),
            ("Thursday", "11/05/2021"), ("Friday", "12/05/2021")]
else:
    # 6 Days FOR EXAMS
    days = [("Monday", "01/05/2021"), ("Tuesday", "02/05/2021"), ("Wednesday", "03/05/2021"),
            ("Thursday", "04/05/2021"),
            ("Friday", "05/05/2021"), ("Monday", "08/05/2021")]

# Exams Time
times = ["09:00 - 12:00", "02:00 - 05:00"]

"""------------------------------------Classes Used In The Code-------------------------------------------"""


"""---Teacher Class-----"""


# Reading Teachers Data from given csv
class Teacher:
    def __init__(self, tid, name):
        self.tid = tid
        self.name = name
        self.binary_val = bin(tid)[2:].zfill(6)

    def __repr__(self):
        return '({0},{1},{2})\n'.format(self.name, self.tid, self.binary_val)


teachers_csv = pd.read_csv(folder + 'teachers.csv', header=None, names=['teachers'])
temp_index = teachers_csv.index.tolist()
temp_teachers_list = teachers_csv.teachers.tolist()
index = []
teachers_list = []
for i in range(len(temp_teachers_list)):
    if temp_teachers_list[i] not in teachers_list:
        teachers_list.append(temp_teachers_list[i])
        index.append(temp_index[i])

teachers = tuple(zip(index, teachers_list))

teachers_data = []
for i in range(len(teachers)):
    x, y = teachers[i]
    teachers_data.append(Teacher(x, y))

# print("\nTEACHERS: \n")
# print(teachers_data)

"""---Room Class-----"""


# Reading Rooms Data - 10 ROOMS
class Room:
    def __init__(self, rid, num, cap):
        self.rid = rid
        self.num = num
        self.cap = cap
        self.binary_val = bin(rid)[2:].zfill(6)

    def __repr__(self):
        return '({0},{1},{2})\n'.format(self.num, self.cap, self.binary_val)


room_csv = pd.read_csv(folder + 'Rooms.csv', header=None, names=['RoomNo', 'Capacity'])
temp_index = room_csv.index.tolist()
room_no = room_csv.RoomNo.tolist()
capacity = room_csv.Capacity.tolist()
rooms = tuple(zip(room_no, capacity))

rooms_data = []
for i in range(len(rooms)):
    x, y = rooms[i]
    rooms_data.append(Room(temp_index[i], x, y))

# print("\nROOMS: (Room No/Capacity)\n")
# print(rooms_data)

"""---Course Class-----"""


# Reading Courses Data from given csv
class Course:
    def __init__(self, cid, code, name):
        self.cid = cid
        self.code = code
        self.name = name
        self.binary_val = bin(cid)[2:].zfill(6)

    def __repr__(self):
        return '({0},{1},{2})\n'.format(self.code, self.name, self.binary_val)


courses_csv = pd.read_csv(folder + 'courses.csv', header=None, names=['course_code', 'course_name'])
temp_index = courses_csv.index.tolist()
temp_code = courses_csv.course_code.tolist()
temp_name = courses_csv.course_name.tolist()

code = []
name = []
for i in range(len(temp_code)):
    if temp_code[i] not in code:
        code.append(temp_code[i])
        name.append(temp_name[i])

courses = tuple(zip(code, name))

courses_data = []
for i in range(len(courses)):
    x, y = courses[i]
    courses_data.append(Course(temp_index[i], x, y))

# print("\nTOTAL COURSES: (CODE/NAME)\n")
# print(courses_data)

"""---Student Class-----"""


# Reading Students Data from given csv
class Student:
    def __init__(self, name, courses):
        self.name = name
        self.courses = courses

    def __repr__(self):
        return '({0},{1})\n'.format(self.name, self.courses)


student_courses_csv = pd.read_csv(folder + 'studentCourse.csv', names=['index', 'student', 'course_code'], skiprows=[0])
temp_name = student_courses_csv.student.tolist()
code = student_courses_csv.course_code.tolist()
temp = tuple(zip(temp_name, code))

for i in range(len(code)):
    code[i] = [code[i]]

name = {}
for n, c in temp:
    if n not in name:
        name[n] = len(name)
        code[name[n]].pop(0)
        if c in temp_code:
            code[name[n]].extend([c])
    else:
        if c not in code[name[n]]:
            if c in temp_code:
                code[name[n]].extend([c])

student_courses = tuple(zip(name, code))

students_data = []
for i in range(len(student_courses)):
    x, y = student_courses[i]
    students_data.append(Student(x, y))

# print("\nSTUDENTS AND THEIR COURSES:\n ")
# print(students_data)


"""---Time Class-----"""


class Time:
    def __init__(self, tid, time):
        self.tid = tid
        self.time = time
        self.binary_val = bin(tid)[2:].zfill(6)

    def __repr__(self):
        return '({0},{1},{2})\n'.format(self.time, self.tid, self.binary_val)

times_data = []
for i in range(len(times)):
    times_data.append(Time(i, times[i]))

# print("\nTIME:\n ")
# print(times_data)

"""---Day Class-----"""


class Day:
    def __init__(self, tid, day):
        self.tid = tid
        self.day = day
        self.binary_val = bin(tid)[2:].zfill(6)

    def __repr__(self):
        return '({0},{1},{2})\n'.format(self.day, self.tid, self.binary_val)

days_data = []
for i in range(len(days)):
    days_data.append(Day(i, days[i]))

# print("\nDays:\n ")
# print(days_data)

"""---Exam Class-----"""


# This Class's 1 instance is referred as a single Chromosome
class Exam:
    def __init__(self, course):
        self.course = course
        self.room = []
        self.time = None
        self.teacher = []
        self.day = None
        self.binary_value = []

    def __repr__(self):
        return '({0}, {1}, {3}, {2}, {4},\n{5})\n'.format(self.course, self.room, self.time, self.teacher, self.day,
                                                          self.binary_value)

    def printBinary(self):
        for i in range(len(self.binary_value)):
            if i == 1 or i == 2:
                print("[ ", end="")
                for j in range(len(self.binary_value[i])):
                    print(*self.binary_value[i][j], sep='', end=" ")
                print("] ", end="")
            else:
                print(*self.binary_value[i], sep='', end=" ")


"""---Population Class-----"""


# It Contains All the Chromosomes
class Population:
    def __init__(self, number, fitness, chromosome):
        self.number = number
        self.fitness = fitness
        self.chromosome = chromosome

    def __repr__(self):
        return ('(Population Number: {0}\nFitness: {1}\nChromosome: \n{2})\n'.format(self.number, self.fitness,
                                                                                     self.chromosome))


"""-------------------------------------CrossOver Function--------------------------------"""


# Function for applying the Crossover To generate Offsprings
def apply_crossover(population):
    j = 1
    crossover_population = []
    for i in range(population_size):
        if random.randint(0, 100) <= crossover_probability * 100:

            # Randomly selecting two chromosomes
            first = random.randint(0, len(population) - 1)
            second = random.randint(0, len(population) - 1)
            # first, second = find_fittest(population)

            crossovered_exams1 = []
            crossovered_exams2 = []

            # Selecting half exams of one chromosome and half of other to make two new chromosomes
            for k in range(len(population[0].chromosome)):
                if k <= int(len(population[0].chromosome) / 2):
                    crossovered_exams1.append(population[first].chromosome[k])
                    crossovered_exams2.append(population[second].chromosome[k])
                else:
                    crossovered_exams1.append(population[second].chromosome[k])
                    crossovered_exams2.append(population[first].chromosome[k])

            new_crossovered_exams1 = apply_mutation(crossovered_exams1.copy())
            new_crossovered_exams2 = apply_mutation(crossovered_exams2.copy())

            fitness = calculateFitness(new_crossovered_exams1, courses_data, students_data, teachers_data)
            x, y = fitness
            fitness = (x, (100 - y))
            crossover_population.append(Population(j, fitness, crossovered_exams1))

            fitness = calculateFitness(new_crossovered_exams2, courses_data, students_data, teachers_data)
            x, y = fitness
            fitness = (x, (100 - y))
            crossover_population.append(Population(j + 1, fitness, crossovered_exams2))

            j += 2

    return crossover_population


"""-------------------------------------Mutation Function--------------------------------"""


# Apply Mutation on chromosomes , given Mutation probability
def apply_mutation(chromosome):
    # Randomly switch values(Day, Time) of a chromosome
    if random.randint(0, 100) <= mutation_probability * 100:
        genes = random.randint(0, 3)
        for i in range(genes):
            gene = random.randint(0, len(chromosome) - 1)
            temp = random.randint(0, len(days_data) - 1)

            # Date Switch
            chromosome[gene].day = days_data[temp].day
            chromosome[gene].binary_value[4] = days_data[temp].binary_val

            # Time Switch
            temp = random.randint(0, len(times_data) - 1)
            chromosome[gene].time = times_data[temp].time
            chromosome[gene].binary_value[3] = times_data[temp].binary_val

            # Room Switch
            temp = random.randint(0, len(rooms_data) - 1)
            while rooms_data[temp].num in chromosome[gene].room:
                temp = random.randint(0, len(rooms_data) - 1)

            temp1 = random.randint(0, len(chromosome[gene].room) - 1)
            chromosome[gene].room[temp1] = rooms_data[temp].num
            chromosome[gene].binary_value[1][temp1] = rooms_data[temp].binary_val

    return chromosome


"""-------------------------------------Parent_Selection Function--------------------------------"""


# Roulette Wheel Selection
# Select  parents from Population
def parent_selection(population):
    parents = []
    total_value = 0  # Total Fitness

    for individual in population:
        total_value += individual.fitness[1]
    # Applying Roulette Wheel
    # Check Total sum value of fittest individuals
    while len(parents) < population_size:
        # Finding  probability of all chromosomes
        probability_arr = []
        for i in range(0, len(population)):
            s = (round((population[i].fitness[1] / total_value), 2), i)
            probability_arr.append(s)

        # sorting the array having the probabilities
        probability_arr.sort(key=lambda val: val[0], reverse=True)
        # Making proper probability pir chart array
        temp_prob_arr = []
        check_temp = 0
        for i in range(0, len(population)):
            prob, ind = probability_arr[i]
            temp_prob_arr.append((prob + check_temp, ind))
            check_temp += prob
        # deciding the chromosome by random number generation and probability
        random_no = round(random.uniform(0, 1), 2)
        for i in range(0, len(temp_prob_arr)):
            prob, ind = temp_prob_arr[i]
            if random_no <= prob:
                parents.append(population[ind])
                break
    return parents


"""-------------------------------------Fitness Calculation Function--------------------------------"""


# This Function Calculates the Fitness of the Chromosome based on the Constraints

def calculateFitness(exams, courses, students, teachers):
    conditions = [0, 0, 0, 0, 0, 0]
    """---Hard Constraint---No exams on weekend----- taken care off"""
    """---Hard Constraint---Exams between 9 am to 5 pm----- taken care off"""

    """---Soft Constraint---Break from 1 pm - 2 pm on Friday----- taken care off"""
    """---Soft Constraint---2 hours break for faculty meetings----- taken care off"""

    """---Soft Constraint Correction---- Checking If MG Courses of each Student before CS"""
    for i in students:
        daysMG = []
        daysCS = []
        ind = 0
        for j in exams:
            if j.course in i.courses and j.course[0:2] == "MG":
                day, date = j.day
                daysMG.append((int(date[0:2]), int(j.time[1]), ind))
            if j.course in i.courses and j.course[0:2] == "CS":
                day, date = j.day
                daysCS.append((int(date[0:2]), int(j.time[1]), ind))
            ind += 1
        daysMG.sort(key=lambda val: val[0])
        daysCS.sort(key=lambda val: val[0])
        for k in range(0, len(daysMG)):
            day, time, indx = daysMG[k]
            for l in range(0, len(daysCS)):
                day2, time2, indx2 = daysCS[l]
                if (day == day2 and time == 2 and time2 == 9) or day > day2:
                    check = exams[indx].day
                    exams[indx].day = exams[indx2].day
                    exams[indx2].day = check
                    break
    """---Soft Constraint---- Checking If MG Courses of each Student before CS"""

    for i in students:
        daysMG = []
        daysCS = []
        for j in exams:
            if j.course in i.courses and j.course[0:2] == "MG":
                day, date = j.day
                daysMG.append((int(date[0:2]), int(j.time[1])))
            if j.course in i.courses and j.course[0:2] == "CS":
                day, date = j.day
                daysCS.append((int(date[0:2]), int(j.time[1])))
        daysMG.sort(key=lambda val: val[0])
        daysCS.sort(key=lambda val: val[0])
        for k in range(0, len(daysMG)):
            day, time = daysMG[k]
            for l in range(0, len(daysCS)):
                day2, time2 = daysCS[l]
                if (day == day2 and time == 2 and time2 == 9) or day > day2:
                    conditions[0] += 1
                    break
    """---Soft Constraint---- Checking For Students with Consecutive"""

    '''
    for i in students:
        days = []
        for j in exams:
            if j.course in i.courses:
                if j.day not in days:
                    days.append(j.day)
                else:
                    conditions[1] += 1
    '''

    """ ---Hard Constraint--- Checking IF all Courses Were Scheduled or Not"""

    for i in courses:
        check = False
        for j in exams:
            if i.code == j.course:
                check = True
        if not check:
            conditions[2] += 1

    """ ----Hard Constraint------ Checking For Students with Exams at the same time"""
    for i in students:
        days = []
        for j in exams:
            if j.course in i.courses:
                if (j.day, j.time) not in days:
                    days.append((j.day, j.time))
                    # print("Correct: ",j.course,i.name,j.day,j.time)
                else:
                    conditions[3] += 1
                    # print("Incorrect: ",j.course,i.name,j.day,j.time)

    """ ---- 2 Hard Constraints---Checking For Teachers with consecutive or at the same time exam duty"""
    for i in teachers:
        days2 = []
        for k in range(0, len(exams)):
            if i.name in exams[k].teacher:
                if exams[k].day not in days2:
                    days2.append(exams[k].day)
                else:
                    conditions[4] += 1

    """----Self Constraint----Checking if same rooms are assigned to two courses at the same time"""
    for i in range(0, len(exams)):
        for j in exams[i].room:
            for k in range(0, len(exams)):
                if i != k and j in exams[k].room and exams[i].day == exams[k].day and exams[i].time == exams[k].time:
                    conditions[5] += 1

    return conditions, conditions[0] + conditions[1] + conditions[2] + conditions[3] + conditions[4] + conditions[5]


"""-------------------------------------Find Fittest Function--------------------------------"""


# As the Name Suggests this function Finds the Two Fittest Chromsomes From the Population

def find_fittest(population):
    population.sort(key=lambda val: val.fitness[1], reverse=True)

    return population[0],population[1]


"""-------------------------------------Generate_Population Function--------------------------------"""


# This Function Makes the chromosomes Randomly Based on the Size of Population Given to It. (Initial Population)

def generate_population(population_size):
    population = []
    for j in range(population_size):
        exams = []
        t_num = 0
        # Assigning random values to each course
        for i in range(len(courses)):
            room_bin = []
            teachers_bin = []

            # Assign Course
            exam = Exam(courses_data[i].code)

            # Assign Room
            room_no = random.randint(0, len(rooms_data) - 1)
            exam.room.append(rooms_data[room_no].num)
            room_bin.append(rooms_data[room_no].binary_val)

            # Assign Time
            temptime = random.randint(0, len(times_data) - 1)
            exam.time = times_data[temptime].time

            # Assign Teacher
            exam.teacher.append(teachers_data[t_num].name)
            teachers_bin.append(teachers_data[t_num].binary_val)

            # Assign Day
            tempday = random.randint(0, len(days_data) - 1)
            exam.day = days_data[tempday].day

            # Getting total number of students taking a course
            students_num = 0
            for num in range(len(students_data)):
                if courses_data[i].code in students_data[num].courses:
                    students_num += 1

            t_num += 1
            if t_num == len(teachers_data) - 1:
                t_num = 0
            # Check if students are more than room capacity
            # Assign different room and teachers
            while students_num > rooms_data[room_no].cap:

                students_num -= rooms_data[room_no].cap

                temp_room_no = random.randint(0, len(rooms_data) - 1)
                while rooms_data[temp_room_no].num in exam.room:
                    temp_room_no = random.randint(0, len(rooms_data) - 1)

                exam.room.append(rooms_data[temp_room_no].num)
                room_bin.append(rooms_data[temp_room_no].binary_val)

                exam.teacher.append(teachers_data[t_num].name)
                teachers_bin.append(teachers_data[t_num].binary_val)

                t_num += 1
                if t_num == len(teachers_data) - 1:
                    t_num = 0

            # Binary Chromosome
            exam.binary_value.append(courses_data[i].binary_val)
            exam.binary_value.append(room_bin)
            exam.binary_value.append(teachers_bin)
            exam.binary_value.append(times_data[temptime].binary_val)
            exam.binary_value.append(days_data[tempday].binary_val)

            exams.append(exam)

        fitness = calculateFitness(exams, courses_data, students_data, teachers_data)
        x, y = fitness
        fitness = (x, (100 - y))
        population.append(Population(j + 1, fitness, exams))

    return population


"""-------------------------------------Display Function--------------------------------"""
"""Type--1"""


# A function to Print the Exam Schedule
def displaySchedule(schedule):
    print("-------------------------------Examination Schedule--------------------------")
    print("Course\tRoom No \t   Day\t\t   Date\t\t   Timing\t\tInvigilator ")
    days22 = []
    count = 0
    for i in schedule.chromosome:
        day, date = i.day
        days22.append((date, int(i.time[0:2]), count))
        count += 1
    days22.sort(key=lambda val: val[1], reverse=True)
    days22.sort(key=lambda val: val[0])

    for i in range(len(days22)):
        print(schedule.chromosome[days22[i][2]].course, end="\t")
        for j in schedule.chromosome[days22[i][2]].room:
            print(j, end=" ")
        for j in range(3, len(schedule.chromosome[days22[i][2]].room), -1):
            print("\t", end="")
        day, date = schedule.chromosome[days22[i][2]].day
        if day == "Friday" or day == "Monday":
            print(day, end="\t\t")
        else:
            print(day, "\t", end="")
        print(date, "\t", schedule.chromosome[days22[i][2]].time, "\t", schedule.chromosome[days22[i][2]].teacher,
              "\t", )
    arr, Fitness = schedule.fitness
    print("\nFitness :: ", Fitness, "/ 100")


"""----------------------------------Display Fitness-----------------------------"""

"""----------------------------------Display Fitness-----------------------------"""


def displayFitness(Fitness):
    conditions, fitness = Fitness
    print("Course Schedule Violations: ", end="")
    print(conditions[2], end="  ")
    print("Same Time Exam[Student] Violations: ", end="")
    print(conditions[3], end="  ")
    print("Same Time or Consecutive Duty[Teacher] Violations: ", end="")
    print(conditions[4], end="  ")
    print("Self_Cond--Same Room Violations: ", end="")
    print(conditions[5], end="  ")
    print("MG Courses before CS Violations: ", end="")
    print(conditions[0])

    print("Total Fitness = ", fitness, " / 100")


"""----------------------------------Display Constraints-----------------------------"""


# A Function TO Display Constraints Followed
def display_Constraints(fitness):
    conditions, fit = fitness
    """---Hard Constraint---No exams on weekend----- taken care off"""
    """---Hard Constraint---Exams between 9 am to 5 pm----- taken care off"""

    """---Soft Constraint---Break from 1 pm - 2 pm on Friday----- taken care off"""
    """---Soft Constraint---2 hours break for faculty meetings----- taken care off"""
    print("        [Constraint]                       [Status]")
    print("[Hard] No exams on weekend                [Followed]")
    print("[Hard] Exams between 9 am to 5 pm         [Followed]")

    print("[Hard] All Courses Were Scheduled         ", end="")
    if conditions[2] == 0:
        print("[Followed]")
    else:
        print("[Violated]")

    print("[Hard] No Student With Exam At Same Time  ", end="")
    if conditions[3] == 0:
        print("[Followed]")
    else:
        print("[Violated]")

    print("[Hard] Teacher With Two Duties Same Time  ", end="")
    if conditions[4] == 0:
        print("[Followed]")
    else:
        print("[Violated]")

    print("[Hard] Teacher With Consecutive Duties    ", end="")
    if conditions[4] == 0:
        print("[Followed]")
    else:
        print("[Violated]")

    print("[Hard] Exam in Same Room[Same Time]       ", end="")
    if conditions[5] == 0:
        print("[Followed]")
    else:
        print("[Violated]")

    print("[Soft] Break from 1 pm - 2 pm on Friday   [Followed]")
    print("[Soft] 2 hours break for faculty meetings [Followed]")
    print("[Soft] Students with Consecutive Exams    [Followed]")

    print("[Soft] MG Courses before CS[Each Student] ", end="")
    if conditions[0] == 0:
        print("[Followed]")
    else:
        print("[Violated]")


def genetic_algo():
    best_solution = None
    max_iter = 100
    population = generate_population(population_size)
    # For Generation Count
    count = 0
    regen = 1
    # Loop for The Generations

    for i in range(max_iter):

        population1 = apply_crossover(population.copy())
        population = parent_selection(population1.copy())
        temp_best, _ = find_fittest(population.copy())
        if best_solution is None:
            best_solution = temp_best
        elif temp_best.fitness < best_solution.fitness:
            best_solution = temp_best
        elif best_solution.fitness[1] > 90:
            regen += 1

        if best_solution.fitness[1] >= 100:
            break

        if i % 10 == 0:
            print("Generation no : ", i)
            displayFitness(best_solution.fitness)
        count += 1

        # If stuck on same fitness for long time, re-generate.

        if regen % 25 == 0:
            regen = 0
            population = generate_population(population_size)


    displaySchedule(best_solution)
    print("Generation =", count)
    print()
    display_Constraints(best_solution.fitness)

    '''
    print("\nChromosomes Values Order: \nCourse\t    [Room]\t\t    [Teacher]\t\tTime   Day")
    for i in range(len(best_solution.chromosome)):
        best_solution.chromosome[i].printBinary()
        print()
    '''


genetic_algo()
```
